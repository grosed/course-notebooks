## Background

### What is Python ?

Python is a program language. As such, it can be used to program a computer to undertake a well defined set of actions. However, a Python program cannot do this on its own - after all, you could write the Python program using pencil and paper !!. To do so, it needs a Python system. A Python system can convert a Python program into a set of instructions that a particular (digital) computer can interpret. There are many different Python systems. 

### What is Data Manipulation ?

Data manipulation is the process of modifying data to make it easier to understand, read, or organize.

There are many software tools available that are designed specifically for data manipulation. For example, a data base system is, to a great extent, a large data manipulation system. Other examples include spreadsheets, statistical software packages, such as Minitab and Genstat, and data visualistion systems such as tableau. 

### What is Data Analysis ?

Data analysis is a process of modeling data with the goal of discovering useful information.

Many data manipulation tools also do some data analysis. This reflects the fact that data manipulation is often undertaken to allow modelling to take place.

### Ordinary least squares regression with statsmodels

This example looks at regressing population against industry. When using __statsmodels__, a model is created first.

```python
import statsmodels.formula.api as smf
```

```python
pollute_model = smf.ols('Population ~ Industry', data=pollute_data)
```

```python
help(pollute_model)
```

First note the way in which the model is specified using the string 'Population ~ Industry'. Secondly, the model uses the formula to create various things, such as the design matrix, but does not actually fit the model. To see what information the model does contain the python __dir__ function can be used.

```python
dir(pollute_model)
```

One of the methods, __fit__, fits the model to the data.

```python
pollute_fit = pollute_model.fit()
```

### Hypothesis testing with SciPy

The __statsmodels__ module is very useful for modelling data. However, for hypothesis testing, the __scipy__ module is usually preferred.

#### Kolmogorov-Smirnov tests

The wings data set contains count data for the size of insects wings from two different locations.

```python
wings_data
```

This example tests the hypothesis that the distribution of wing sizes from the two locations are the same.

```python
import scipy
```

```python
res = scipy.stats.kstest(wings_data['LocationA'],wings_data['LocationB'])
print(res)
```

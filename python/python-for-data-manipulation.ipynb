## What is Python ?

Python is a program language. As such, it can be used to program a computer to undertake a well defined set of actions. However, a Python program cannot do this on its own - after all, you could write the Python program using pencil and paper !!. To do so, it needs a Python system. A Python system can convert a Python program into a set of instructions that a particular (digital) computer can interpret. There are many different Python systems. 

### What is Data Manipulation ?

Data manipulation is the process of modifying data to make it easier to understand, read, or organize.

There are many software tools available that are designed specifically for data manipulation. For example, a data base system is, to a great extent, a large data manipulation system. Other examples include spreadsheets, statistical software packages, such as Minitab and Genstat, and data visualistion systems such as tableau. 

### What is Data Analysis ?

Data analysis is a process of modeling data with the goal of discovering useful information.

Many data manipulation tools also do some data analysis. This reflects the fact that data manipulation is often undertaken to allow modelling to take place.

#### Accesing data in a pandas data frame

There are many ways of accessing and subsetting the data in a data frame. The follwing examples show some of the more commonly used methods.

##### By column name

```python
pollute_data[['Temp','Wind']]
```

##### By row number

```python
pollute_data.iloc[2:7]
```

##### By row number and column name

```python
pollute_data[2:7][['Temp','Rain']]
```

##### By row number and column number

```python
pollute_data.iloc[2:7][3:7]
```

```python
pollute_data.iloc[[1,4,5]][1:3]
```

##### Using a logical condition

```python
pollute_data[pollute_data['Wind'] > 10.0]
```
